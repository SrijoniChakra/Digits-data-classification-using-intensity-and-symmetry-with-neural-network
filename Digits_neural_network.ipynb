```python
def compute_intensity(X):
    intensity = np.mean(X, axis=1)
    return intensity
```

```python
def compute_symmetry(X,img_width,img_height):
    N, d = X.shape
    Xflip = X.reshape(N, 16, 16)
    Xflip = np.flip(Xflip, axis=2)
    Xflip = Xflip.reshape(N, d)
    asym = np.abs(X - Xflip)
    asym = np.mean(asym, axis = 1)
    sym = -asym
    return sym
```

```python
input_data=train_features.loc[:, train_features.columns != 'x0']
input_data=input_data.to_numpy()
input_data[:,-1].shape
```

```python
#for 2 input features and 10 hidden units, the weights to layer 1 should have dimension (2,10) and weights to output layer 
#should be (10,1) to get target y of dimesion (1561,1)  
def weights():
    W_layer1 = np.random.randn(2,10)
    W_layer2 = np.random.randn(10,1) 
    return W_layer1,W_layer2
```

```python
def bias(input_data):
    bshape=len(input_data)
    b1=np.zeros((bshape, 10))
    b2=np.zeros((bshape, 1))
    return b1,b2
```

```python
def tanh(z):
    return (np.exp(z)-np.exp(-z))/(np.exp(z)+np.exp(-z))
```

```python
def tanh_der(z):
    return (1-(tanh(z)*tanh(z)))

def sign(data):
    data[data < 0] = -1
    data[data >= 0] = 1
    return data
```

```python
def feedforward(input_features,W_layer1,W_layer2,b1,b2):
    input_layer1= np.dot(input_features[:,[0,1]],W_layer1)+b1
    output_layer1=tanh(input_layer1)
    input_layer2=np.dot(output_layer1,W_layer2)+b2
    output_layer2=tanh(input_layer2)
    return input_layer1,output_layer1,input_layer2,output_layer2

def backprop(input_features,W_layer1,W_layer2,input_layer1,output_layer1,input_layer2,output_layer2):
    tshape= input_features.shape[0] 
    target=input_features[:,-1].reshape((tshape, 1))
    error_out=((1/2)*(np.power((output_layer2 - target),2)))   #mean squared error 
    error_sum=error_out.sum()
    de_dout2=2*(output_layer2-target)
    dout2_din2=tanh_der(input_layer2)
    din2_dw2=output_layer1
    de_dw2= np.dot(din2_dw2.T,de_dout2*dout2_din2)  #x(1).del(2)
    de_din2=de_dout2*dout2_din2 
    din2_dout1=W_layer2
    de_dout1=np.dot(de_din2,din2_dout1.T) 
    dout1_din1=tanh_der(input_layer1) 
    del1=dout1_din1*de_dout1
    din1_dw1=input_features[:,[0,1]]
    de_dw1=np.dot(din1_dw1.T,del1)  #x(0).del(1)
    return de_dw1,de_dw2,error_sum,del1,de_din2
```

```python
#weights initialization for gradient descent
W_layer1,W_layer2=weights()
```

```python
#neural network training using gradient descent 
error_gd=[]
b1,b2 = bias(input_data)
for epoch in range(1000):
    input_layer1,output_layer1,input_layer2,output_layer2=feedforward(input_data,W_layer1,W_layer2,b1,b2)
    de_dw1,de_dw2,error_sum,del1,de_din2=backprop(input_data,W_layer1,W_layer2,input_layer1,output_layer1,input_layer2,output_layer2)
    error_gd.append(error_sum)
    lr=0.001
    W_layer1-=lr*de_dw1
    W_layer2-=lr*de_dw2
    b1-=lr*del1
    b2-=lr*de_din2
```

```python
plt.plot(error_gd)
plt.title('Gradient descent')
plt.ylabel('log error')
plt.xlabel('log epoch')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
#weights initialization for mini-batch gradient descent
W_layer1_mbgd,W_layer2_mbgd=weights()
```

```python
#neural network training using mini-batch gradient descent 
n = len(input_data)
batch_size=32
n_mini_batches = input_data.shape[0]
error_mbgd=[]
for j in range(1000):
    np.random.shuffle(input_data)
    mini_batches = [input_data[k:k+batch_size] for k in range(0, n, batch_size)]
    for mini_batch in mini_batches:
        b1_mbgd,b2_mbgd = bias(mini_batch)
        input_l1_mbgd,output_l1_mbgd,input_l2_mbgd,output_l2_mbgd=feedforward(mini_batch,W_layer1_mbgd,W_layer2_mbgd,b1_mbgd,b2_mbgd)
        de_dw1_mbgd,de_dw2_mbgd,error_sum_mbgd,del1,de_din2=backprop(mini_batch,W_layer1_mbgd,W_layer2_mbgd,input_l1_mbgd,output_l1_mbgd,input_l2_mbgd,output_l2_mbgd)
        error_mbgd.append(error_sum_mbgd)
        lr=0.001
        W_layer1_mbgd-=lr*de_dw1_mbgd
        W_layer2_mbgd-=lr*de_dw2_mbgd
        b1_mbgd-=lr*del1
        b2_mbgd-=lr*de_din2
```

```python
plt.plot(error_mbgd)
plt.title('Mini-batch Gradient descent')
plt.ylabel('log error')
plt.xlabel('log epoch')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
#weights initialization for stochastic gradient descent
W_layer1_sgd,W_layer2_sgd=weights()
```

```python
#neural network training using stochastic gradient descent 
n = len(input_data)
batch_size=1
n_mini_batches = input_data.shape[0]
error_sgd=[]
for j in range(1000):
    np.random.shuffle(input_data)
    mini_batches = [input_data[k:k+batch_size] for k in range(0, n, batch_size)]
    for mini_batch in mini_batches:
        b1_sgd,b2_sgd = bias(mini_batch)
        input_l1_sgd,output_l1_sgd,input_l2_sgd,output_l2_sgd=feedforward(mini_batch,W_layer1_sgd,W_layer2_sgd,b1_sgd,b2_sgd)
        de_dw1_sgd,de_dw2_sgd,error_sum_sgd,del1,de_din2=backprop(mini_batch,W_layer1_sgd,W_layer2_sgd,input_l1_sgd,output_l1_sgd,input_l2_sgd,output_l2_sgd)
        error_sgd.append(error_sum_sgd)
        lr=0.001
        W_layer1_sgd-=lr*de_dw1_sgd
        W_layer2_sgd-=lr*de_dw2_sgd
        b1_sgd-=lr*del1
        b2_sgd-=lr*de_din2
```

```python
plt.plot(error_sgd)
plt.title('Stochastic Gradient descent')
plt.ylabel('log error')
plt.xlabel('log epoch')
plt.xscale('log')
plt.yscale('log')
plt.show()
```
